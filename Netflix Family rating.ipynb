```python
import pandas as pd
```

**Reading file**

```python
df = pd.read_csv(r'/kaggle/input/netflix-shows/netflix_titles.csv')
df
```

**Displaying info**

```python
df.info()
```

**Determine Shows inappropriate for family**

```python
df2 = df[df['rating'].str.contains('R',na=False)]
df2
```

**Deleting inappropriate shows**

```python
for x in df.index:
    if df.loc[x,'rating'] == 'R' or df.loc[x,'rating'] == 'NR' or df.loc[x,'rating'] == 'UR' or df.loc[x,'rating'] == 'NC-17':
        df.drop(x,inplace=True)
df
```

**Renaming rating column**

```python
df.rename(columns={'rating':'rate'}, inplace=True)
df
```
